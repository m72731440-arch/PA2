```python
# Solicitando o número ao usuário
# Usamos int() para converter o texto digitado em um número inteiro
numero = int(input("Digite um número inteiro: "))

# Verificando se o número é par ou ímpar
if numero % 2 == 0:
    print(f"O número {numero} é PAR.")
else:
    print(f"O número {numero} é ÍMPAR.")

    # Solicitando dois números ao usuário
# Usamos float() para permitir que o usuário digite números decimais (ex: 5.5)
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Comparando os valores
if num1 > num2:
    print(f"O primeiro número ({num1}) é maior que o segundo ({num2}).")
elif num2 > num1:
    print(f"O segundo número ({num2}) é maior que o primeiro ({num1}).")
else:
    print("Os dois números são iguais.")

    # Solicitando a nota do aluno
# Usamos float() pois notas geralmente possuem casas decimais
nota = float(input("Digite a nota do aluno (0 a 10): "))

# Verificando a situação do aluno
if nota >= 7:
    print("Situação: APROVADO")
elif 5 <= nota < 7:
    print("Situação: RECUPERAÇÃO")
else:
    print("Situação: REPROVADO")


  # Solicitando a idade do usuário
idade = int(input("Digite a idade: "))

# Classificando a faixa etária
if idade < 0:
    print("Idade inválida!")
elif idade <= 12:
    print("Classificação: Criança")
elif idade <= 17:
    print("Classificação: Adolescente")
elif idade <= 59:
    print("Classificação: Adulto")
else:
    print("Classificação: Idoso")


    # Definindo a senha secreta (senha fixa)
SENHA_CORRETA = "python123"

# Solicitando que o usuário digite a senha
tentativa = input("Digite a senha para acessar o sistema: ")

# Verificando se a tentativa coincide com a senha fixa
if tentativa == SENHA_CORRETA:
    print("Acesso Permitido! Bem-vindo.")
else:
    print("Acesso Negado! Senha incorreta.")


# Solicitando o número ao usuário
# Usamos float() para que o programa aceite números como -5.5 ou 10
numero = float(input("Digite um número: "))

# Verificando a natureza do número
if numero > 0:
    print(f"O número {numero} é POSITIVO.")
elif numero < 0:
    print(f"O número {numero} é NEGATIVO.")
else:
    print("O número é ZERO.")


# Solicitando os números
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Solicitando a operação
print("Operações disponíveis: +, -, *, /")
operacao = input("Qual operação deseja realizar? ")

# Lógica da calculadora
if operacao == "+":
    resultado = num1 + num2
    print(f"Resultado: {num1} + {num2} = {resultado}")

elif operacao == "-":
    resultado = num1 - num2
    print(f"Resultado: {num1} - {num2} = {resultado}")

elif operacao == "*":
    resultado = num1 * num2
    print(f"Resultado: {num1} * {num2} = {resultado}")

elif operacao == "/":
    # Verificação de segurança: não existe divisão por zero!
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado: {num1} / {num2} = {resultado}")
    else:
        print("Erro: Não é possível dividir por zero.")

else:
    print("Operação inválida! Por favor, escolha +, -, * ou /.")



    # Solicitando um número ao usuário
dia_numero = int(input("Digite um número de 1 a 7: "))

# Verificando qual é o dia da semana correspondente
if dia_numero == 1:
    print("O dia correspondente é: Domingo")
elif dia_numero == 2:
    print("O dia correspondente é: Segunda-feira")
elif dia_numero == 3:
    print("O dia correspondente é: Terça-feira")
elif dia_numero == 4:
    print("O dia correspondente é: Quarta-feira")
elif dia_numero == 5:
    print("O dia correspondente é: Quinta-feira")
elif dia_numero == 6:
    print("O dia correspondente é: Sexta-feira")
elif dia_numero == 7:
    print("O dia correspondente é: Sábado")
else:
    print("Número inválido! Por favor, digite um número entre 1 e 7.")


# Solicitando o ano
ano = int(input("Digite um ano: "))

# Verificando se o ano é bissexto
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é BISSEXTO.")
else:
    print(f"O ano {ano} NÃO É BISSEXTO.")


    # Solicitando o valor da compra
valor_original = float(input("Digite o valor da compra (R$): "))

# Aplicando as regras de desconto
if valor_original >= 100:
    desconto = 0.10  # 10%
elif valor_original >= 50:
    desconto = 0.05  # 5%
else:
    desconto = 0.00  # Sem desconto

# Calculando os valores finais
valor_desconto = valor_original * desconto
valor_final = valor_original - valor_desconto

# Exibindo os resultados
print(f"\nValor original: R$ {valor_original:.2f}")
print(f"Desconto aplicado: {desconto * 100:.0f}%")
print(f"Valor do desconto: R$ {valor_desconto:.2f}")
print(f"Total a pagar: R$ {valor_final:.2f}")
```
